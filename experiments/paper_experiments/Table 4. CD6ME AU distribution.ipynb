```python
from meb import datasets
from meb import utils
```

```python
df = datasets.CrossDataset(ignore_validation=True).data_frame
```

```python
# Table of aus for paper
action_units = utils.dataset_aus["cross"]
au_sum_df = df.groupby("dataset")[action_units].sum()
au_sum_df["Total"] = au_sum_df.sum(1)
au_sum_df.loc["Total"] = au_sum_df.sum(numeric_only=True)
au_sum_df
```
